AYUDA: La direccin del banco nos proporciona estas notas (en ingls):

 
   **bank client data:**  
   1 - age (numeric)  
   2 - job : type of job (categorical: "admin.","unknown","unemployed","management","housemaid","entrepreneur","student",
                                       "blue-collar","self-employed","retired","technician","services")   
   3 - marital : marital status (categorical: "married","divorced","single"; note: "divorced" means divorced or widowed)  
   4 - education (categorical: "unknown","secondary","primary","tertiary")  
   5 - default: has credit in default? (binary: "yes","no")  
   6 - balance: average yearly balance, in euros (numeric)   
   7 - housing: has housing loan? (binary: "yes","no")  
   8 - loan: has personal loan? (binary: "yes","no")   

   **related with the last contact of the current campaign:**  
   9 - contact: contact communication type (categorical: "unknown","telephone","cellular")   
  10 - day: last contact day of the month (numeric)  
  11 - month: last contact month of year (categorical: "jan", "feb", "mar", ..., "nov", "dec")  
  12 - duration: last contact duration, in seconds (numeric)    

   **other attributes:**  
  13 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)  
  14 - pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric, -1 means client was not previously contacted)  
  15 - previous: number of contacts performed before this campaign and for this client (numeric)  
  16 - poutcome: outcome of the previous marketing campaign (categorical: "unknown","other","failure","success")

### 4. Tratamiento de variables

```python
categories = [['no','yes'],
              ["admin.","unknown","unemployed","management","housemaid","entrepreneur","student", "blue-collar","self-employed","retired","technician","services"],
              ['no','yes'],
              ['cellular','telephone','unknown'],
              ['jan', 'feb', 'mar', 'apr', 'may', 'jun','jul', 'aug', 'sep','oct', 'nov', 'dec']]
```

```python
encoder = OrdinalEncoder(categories= categories) 
```

#### Procesamiento de categricas y target

```python
train_set[['y','job', 'housing', 'contact', 'month']] = encoder.fit_transform(train_set[['y','job', 'housing', 'contact', 'month']])
test_set[['y','job', 'housing', 'contact', 'month']] = encoder.transform(test_set[['y','job', 'housing', 'contact', 'month']])
```

A la variable pdays le hacemos sqrt.

A la variable previous le hacemos sqrt

```python
# Aplicamos log a day
train_set["day"] = train_set["day"].apply(np.log)
test_set["day"] = test_set["day"].apply(np.log)

# Aplicamos sqrt a duration, pdays y previous
train_set["duration"] = np.sqrt(train_set["duration"])
test_set["duration"] = np.sqrt(test_set["duration"])

train_set["pdays"] = np.sqrt(train_set["pdays"])
test_set["pdays"] = np.sqrt(test_set["pdays"])

train_set["previous"] = np.sqrt(train_set["previous"])
test_set["previous"] = np.sqrt(test_set["previous"])
```

#### Escalamos

```python
train_set[features_num_2].describe()
```

```python
train_set[features_num_2] = scaler.fit_transform(train_set[features_num_2])
```

### 5. Oversampling en Train

```python
X_train = train_set[features_num_2+features_cat_2]
y_train = train_set[target]

X_test = test_set[features_num_2+features_cat_2]
y_test = test_set[target]
```

```python
y_train.value_counts(normalize = True)
```

```python
# reajustamos con oversampling
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

### 6. Modelado

#### A. rbol de decisin

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(max_depth=None,
                                  random_state=42)
```

```python
#entrenar rbol
tree_clf.fit(X_train, y_train)

#predicciones y
y_pred = tree_clf.predict(X_test)

#evaluar contra test
evaluacion = classification_report(y_test, y_pred)
print(evaluacion)
```

#### B. Regresin logstica

```python
# regresion logistica para clasificacin
clf = LogisticRegression(max_iter = 10000)
```

```python
clf.fit(X_train, y_train)
```

```python
print(classification_report(y_test, clf.predict(X_test)))
```

Comparando ambos modelos sin hiperparmetros, el modelo del rbol de decisin es mejor debido al F-Score y al recall.

####  C. rbol de decisin con hiperparmetros

#### Random Search

```python
param_grid_decision_tree = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
random_search_tree = RandomizedSearchCV(tree_clf, # El modelo que quiero entrenar
                                   param_grid_decision_tree, # El grid de parmetros
                                   random_state=1, # Semilla
                                   n_iter=9, # N de iteraciones del random search --> combinaciones que quiero porbar
                                   cv=5, # Folds del cross validation
                                   scoring="accuracy", # Mtrica que quiero optimizar
                                   n_jobs = -1 # Paralelizar
                                  )

random_search_tree.fit(X_train, y_train)
```

```python
# Mejor score (R2) rbol decisin
random_search_tree.best_score_**2
```

```python
# Mejores parmetros rbol decision
random_search_tree.best_params_
```

```python
#modelo con los mejores parmetros
tree_clf_2 = DecisionTreeClassifier(max_depth=10, min_samples_split = 2, min_samples_leaf = 2,
                                  random_state=42)
```

```python
#entrenar rbol
tree_clf_2.fit(X_train, y_train)

#predicciones y
y_pred = tree_clf_2.predict(X_test)

#evaluar contra test
evaluacion = classification_report(y_test, y_pred)
print(evaluacion)
```

#### Grid Search

```python
grid_search_tree = GridSearchCV(tree_clf,
                           param_grid_decision_tree,
                           cv=5,
                           scoring='accuracy',
                           n_jobs=-1
                          )

grid_search_tree.fit(X_train, y_train)
```

```python
print(classification_report(y_test, grid_search_tree.predict(X_test)))
```

Por lo tanto el mejor modelo obtenido ha sido el optimizado con Random Search ya que el presenta el recall ms alto y es la mtrica que necesito maximizar porque es la que me me permite identificar ms compradores.
